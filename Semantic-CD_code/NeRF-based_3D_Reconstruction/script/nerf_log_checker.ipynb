```python
from tensorboard.backend.event_processing import event_accumulator

log_file = "/workspace/Laboratory/02.Rapid3DReconstruction/00.workspace/Naju_90m_ChangeDetection_workspace/initial/nerf_output/ns_processed/nerfacto/2025-03-07_015048/events.out.tfevents.1741312331.9cc87d9b32ff.1426819.0"

ea = event_accumulator.EventAccumulator(log_file)
ea.Reload()  # 로그 데이터 로드

# 가능한 키(예: Scalars) 확인
print(ea.Tags())

# 특정 스칼라 메트릭 출력 (예: loss, accuracy 등)
if 'scalars' in ea.Tags():
    for tag in ea.Tags()['scalars']:
        events = ea.Scalars(tag)
        print(f"Metric: {tag}")
        if(tag == "Eval Metrics Dict/psnr"):
            for event in events:  # 처음 5개 데이터만 출력
                print(f"Step {event.step}: {event.value}")
```

```python
import csv
from tensorboard.backend.event_processing import event_accumulator

log_file = "/workspace/Laboratory/02.Rapid3DReconstruction/00.workspace/Naju_90m_ChangeDetection_workspace/initial/nerf_output/ns_processed/nerfacto/2025-03-07_015048/events.out.tfevents.1741312331.9cc87d9b32ff.1426819.0"

ea = event_accumulator.EventAccumulator(log_file)
ea.Reload()

csv_filename = "/workspace/Laboratory/02.Rapid3DReconstruction/00.workspace/Naju_90m_ChangeDetection_workspace/initial/nerf_output/ns_processed/nerfacto/2025-03-07_015048/training_log.csv"

with open(csv_filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Metric", "Step", "Value"])  # 헤더 작성

    for tag in ea.Tags().get('scalars', []):
        for event in ea.Scalars(tag):
            writer.writerow([tag, event.step, event.value])

print(f"로그 데이터가 {csv_filename} 파일로 저장되었습니다.")
```
